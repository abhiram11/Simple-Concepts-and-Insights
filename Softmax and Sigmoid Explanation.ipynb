```python
from math import exp
```

```python
# (0,1,2,3,4,5)
def softmax(vector):
    ans = []
    denominator = 0
    for i in vector:
        denominator += exp(i)
    
    for i in vector:
        ans.append(round((exp(i)/denominator), 6))
    return ans
```

```python
softmax([7, -7.5, 10])
```

```python
def sigmoid(vector):
    ans = []
    for i in vector:
        ans.append((1/(1+exp(-(i)))))
    return ans
```

```python
sigmoid([7, 1.5, 10])
```
